```python
from keras.utils import get_file

file_name = 'enwiki-20180920-pages-meta-current2.xml-p30304p88444.bz2'
file = get_file(file_name, 
         f'https://dumps.wikimedia.org/enwiki/20180920/{file_name}')
```

```python
lines = []

import subprocess
import json

for i, l in enumerate(subprocess.Popen(['bzcat'], stdin = open(file), 
                          stdout = subprocess.PIPE).stdout):
    lines.append(l)
    if i > 1e6:
        break
```
